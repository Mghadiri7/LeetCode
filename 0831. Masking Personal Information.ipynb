```python
def maskPII(s):
    if "@" in s:
        s = s.lower()
        user, domain = s.split("@")
        user = user[0] + "*****" + user[-1]
        return user + "@" + domain
    else:
        phone = ""
        for x in s:
            if x.isdigit():
                phone += x
        n = len(phone)
        if n > 10:
            return "+" + "*"*(n-10) + "-***-***-" + phone[-4:]
        return "***-***-" + phone[-4:]           
```

```python
s = "LeetCode@LeetCode.com"
maskPII(s)
```

```python
s = "1(234)567-890"
maskPII(s)
```
