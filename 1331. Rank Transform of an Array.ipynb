```python
def arrayRankTransform(arr):
    lst = sorted(set(arr))
    rank_map = {val: i + 1 for i, val in enumerate(lst)}
    return [rank_map[x] for x in arr]            
```

```python
arr = [37,12,28,9,100,56,80,5,12]
arrayRankTransform(arr)
```

```python
arr = [40,10,20,30]
arrayRankTransform(arr)
```
