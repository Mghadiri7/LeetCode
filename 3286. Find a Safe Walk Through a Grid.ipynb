```python
from collections import deque

def findSafeWalk(grid, health):
    m, n = len(grid), len(grid[0])
    start_damage = grid[0][0]

    if health - start_damage <= 0:
        return False

    min_damage = [[float('inf')]*n for i in range(m)]
    min_damage[0][0] = start_damage

    dq = deque([(0, 0)])
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    while dq:
        r, c = dq.popleft()

        if r == m-1 and c == n-1:
            return health - min_damage[r][c] > 0

        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            if 0 <= nr < m and 0 <= nc < n:
                cost = grid[nr][nc]
                new_damage = min_damage[r][c] + cost

                if health - new_damage <= 0:
                    continue

                if new_damage < min_damage[nr][nc]:
                    min_damage[nr][nc] = new_damage

                    if cost == 0:
                        dq.appendleft((nr, nc))
                    else:
                        dq.append((nr, nc))

    return False
```

```python
grid = [[0,1,1,0,0,0],[1,0,1,0,0,0],[0,1,1,1,0,1],[0,0,1,0,1,0]]
health = 3
findSafeWalk(grid, health)
```

```python
grid = [[1,1,1],[1,0,1],[1,1,1]]
health = 5
findSafeWalk(grid, health)
```
