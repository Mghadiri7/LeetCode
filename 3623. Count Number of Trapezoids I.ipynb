```python
from collections import defaultdict

def countTrapezoids(points):
    num_points = defaultdict(int)
    mod = 10**9 + 7
    ans, total_edge = 0, 0
    for point in points:
        num_points[point[1]] += 1
    for p in num_points.values():
        edge = p * (p-1) // 2
        ans = (ans + edge * total_edge) % mod
        total_edge = (total_edge + edge) % mod
    return ans
```

```python
points = [[1,0],[2,0],[3,0],[2,2],[3,2]]
countTrapezoids(points)
```
