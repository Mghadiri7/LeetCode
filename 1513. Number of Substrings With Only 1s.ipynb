```python
def numSub(s):
    ans = 0
    count1 = 0
    for num in s:
        if num=='0' and count1>0:
            ans += count1 * (count1+1) // 2
            count1 = 0
        elif num=='1':
            count1+=1
    if count1>0:
        ans += count1 * (count1+1) // 2

    return ans
    # return ans % 1000000007
```

```python
s = "0110111"
numSub(s)
```
