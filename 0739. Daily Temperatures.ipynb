## First Solution
### Time Limit Exceeded

```python
def dailyTemperatures(temperatures):
    n = len(temperatures)
    ans = [0] * n
    for i in range(n-1):
        for j in range(i+1, n):
            if temperatures[j]>temperatures[i]:
                ans[i] = j-i
                break
    return ans
```

```python
temperatures = [73,74,75,71,69,72,76,73]
dailyTemperatures(temperatures)
```

## Second Solution
### Without TLE

```python
def dailyTemperatures(temperatures):
    n = len(temperatures)
    ans = [0] * n
    lst = [n] * 102
    for i in range(n-1, -1, -1):
        t = temperatures[i]
        lst[t] = i
        min_days = min(lst[t+1:])
        if min_days < n:
            ans[i] = min_days - i
    return ans
```

```python
temperatures = [73,74,75,71,69,72,76,73]
dailyTemperatures(temperatures)
```

## Third Solution
### Efficient Solution

```python
def dailyTemperatures(temperatures):
    n = len(temperatures)
    ans = [0] * n
    stack = []
    for curr_day, t in enumerate(temperatures):
        while stack and t > temperatures[stack[-1]]:
            prev_day = stack.pop()
            ans[prev_day] = curr_day - prev_day
        stack.append(curr_day)

    return ans
```

```python
temperatures = [73,74,75,71,69,72,76,73]
dailyTemperatures(temperatures)
```
