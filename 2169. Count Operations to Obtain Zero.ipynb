```python
def countOperations(num1, num2):
    ans = 0
    while num1>0 and num2>0:
        counter = 0
        if num1<num2:
            num1, num2 = num2, num1
        counter += num1//num2
        num1 = num1 - (num2*counter)
        ans+=counter
    return ans
```

```python
num1 = 2
num2 = 3
countOperations(num1, num2)
```
