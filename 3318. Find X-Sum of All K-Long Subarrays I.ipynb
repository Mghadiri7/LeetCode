```python
from collections import Counter

def findXSum(nums, k, x):
    n = len(nums)
    res = []
    for i in range(n-k+1):
        window = nums[i:i+k]
        _dict = Counter(window)
        sorted_dict = sorted(_dict.items(), key=lambda a: (-a[1], -a[0]))
        lst = sorted_dict[:x]
        _sum = 0
        for j in range(len(lst)):
            _sum += lst[j][0]*lst[j][1]
        res.append(_sum)
    return res
```

```python
nums = [1,1,2,2,3,4,2,3]
k = 6
x = 2
findXSum(nums, k, x)
```

```python
nums = [9,2,2]
k = 3
x = 3
findXSum(nums, k, x)
```
