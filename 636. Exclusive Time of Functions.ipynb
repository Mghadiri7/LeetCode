```python
def exclusiveTime(n, logs):
    res = [0] * n
    time_prev = 0
    stack = []
    for log in logs:
        f_id, state, time = log.split(":")
        f_id, time = int(f_id), int(time)
        if state=="start":
            if stack:
                res[stack[-1]] += time - time_prev
            stack.append(f_id)
            time_prev = time
        else:
            res[stack.pop()] += time - time_prev + 1
            time_prev = time + 1

    return res
```

```python
n = 2
logs = ["0:start:0","1:start:2","1:end:5","0:end:6"]
exclusiveTime(n, logs)
```
