```python
def getCommon(nums1, nums2):
    n = len(nums1)
    m = len(nums2)
    i, j = 0, 0
    while i<n and j<m:
        if nums1[i] == nums2[j]:
            return nums1[i]
        elif nums1[i] > nums2[j]:
            j += 1
        else:
            i += 1
    return -1
```

```python
nums1 = [1,2,3,6]
nums2 = [2,3,4,5]
getCommon(nums1, nums2)
```

```python
nums1 = [1,2,6,7]
nums2 = [0,3,4,5]
getCommon(nums1, nums2)
```
