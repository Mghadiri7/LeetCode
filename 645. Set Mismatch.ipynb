## First Solution

```python
def findErrorNums(nums):
    n = len(nums)
    lst = [0] * (n+1)
    lst[0] = 1
    ans = []
    for num in nums:
        lst[num] += 1
    for i in range(n+1):
        if lst[i]==2:
            duplicate = i
        if lst[i]==0:
            missing = i
    ans.append(duplicate)
    ans.append(missing)
    return ans
```

```python
nums = [1,2,2,4]
findErrorNums(nums)
```

## Second Solution

```python
def findErrorNums(nums):
    n = len(nums)
    res = []
    _sum = sum(nums)
    sum_set = sum(set(nums))
    res.append(_sum - sum_set)
    res.append(n*(n+1)//2 - sum_set)
    return res
```

```python
nums = [1,2,2,4]
findErrorNums(nums)
```
