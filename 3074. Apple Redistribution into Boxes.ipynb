```python
def minimumBoxes(apple, capacity):
    capacity.sort(reverse = True)
    _sum = sum(apple)
    for i in range(len(capacity)):
        _sum -= capacity[i]
        if _sum < 1:
            return i+1
    return -1
```

```python
apple = [1,3,2]
capacity = [4,3,1,5,2]
minimumBoxes(apple, capacity)
```
