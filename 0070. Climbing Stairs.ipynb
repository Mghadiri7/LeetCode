## First Solution

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def climbStairs(n):
    if n==1:
        return 1
    elif n==2:
        return 2
    else:
        return climbStairs(n-1)+climbStairs(n-2)
```

```python
climbStairs(9)
```

## Second Solution

```python
def climbStairs(n):
    if n<=2:
        return n
    a, b = 1, 2
    for i in range(3, n+1):
        a, b = b, a+b
    return b
```

```python
climbStairs(9)
```
