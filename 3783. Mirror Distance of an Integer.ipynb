## First Solution

```python
def mirrorDistance(n):
    number = n
    reverse = 0
    lst = []
    while n>0:
        lst.append(n%10)
        n //= 10
        
    k = len(lst) - 1
    for num in lst:
        reverse += (num * (10**k))
        k -= 1
    print(reverse)
    return abs(number - reverse)
```

```python
n = 25
mirrorDistance(n)
```

## Second Solution

```python
def mirrorDistance(n):
    return abs(n - int(str(n)[::-1]))
```

```python
n = 25
mirrorDistance(n)
```
