## First Solution

```python
def minOperations(s):
    n = len(s)
    zero = ''
    one = ''
    flag = True
    for i in range(n):
        zero += str(int(not flag))
        one += str(int(flag))
        flag = not flag
    
    diff_zero = sum(1 for a, b in zip(s, zero) if a != b)
    diff_one = sum(1 for a, b in zip(s, one) if a != b)
    return min(diff_zero, diff_one)
```

```python
s = "1111"
minOperations(s)
```

## Second Solution

```python
def minOperations(s):
    n = len(s)
    count = 0
    for i, char in enumerate(s):
        expect = "1" if i%2==0 else "0"
        if char != expect:
            count += 1

    return min(count, n-count)
```

```python
s = "110"
minOperations(s)
```
