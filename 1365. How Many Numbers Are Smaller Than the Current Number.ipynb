```python
from collections import defaultdict

def smallerNumbersThanCurrent(nums):
    n = len(nums)
    lst = nums.copy()
    lst.sort()
    _dict = defaultdict(int)
    for i in range(1, n):
        if lst[i]>lst[i-1]:
            _dict[lst[i]] = i
            
    for i in range(n):
        lst[i] = _dict[nums[i]]
    return lst
```

```python
nums = [8,1,2,2,3]
smallerNumbersThanCurrent(nums)
```
