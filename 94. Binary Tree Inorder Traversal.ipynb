## First Solution (Recursion)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution:
    def inorderTraversal(self, root):
        res = []
        def dfs(node):
            if not node:
                return
            dfs(node.left)
            res.append(node.val)
            dfs(node.right)
        dfs(root)
        return res
```

```python
from collections import deque

def _is_null(x):
    return x is None or x == 'null'

def build_tree(values):
    if not values or _is_null(values[0]):
        return None

    root = TreeNode(values[0])
    q = deque([root])
    i = 1

    while q and i < len(values):
        node = q.popleft()

        if i < len(values) and not _is_null(values[i]):
            node.left = TreeNode(values[i])
            q.append(node.left)
        i += 1

        if i < len(values) and not _is_null(values[i]):
            node.right = TreeNode(values[i])
            q.append(node.right)
        i += 1

    return root
```

```python
values = [1,2,3,4,5,'null',8,'null','null',6,7,9]
root = build_tree(values)
sol = Solution()
print(sol.inorderTraversal(root))
```

## Second Solution (Iterative)

```python
class Solution:
    def inorderTraversal(self, root):
        res = []
        stack = []
        curr = root

        while curr or stack:
            while curr:
                stack.append(curr)
                curr = curr.left

            curr = stack.pop()
            res.append(curr.val)
            curr = curr.right

        return res
```
