```python
def countDigitOne(n):
    i = 1
    ans = 0
    while i<=n:
        divider = i * 10
        full_block = (n // divider) * i
        reminder = n % divider
        partial_block = min(max(reminder-i+1, 0), i)
        ans += full_block + partial_block
        i *= 10
    return ans
```

```python
n = 13
countDigitOne(n)
```
