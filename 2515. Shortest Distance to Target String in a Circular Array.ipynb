```python
def closestTarget(words, target, startIndex):
    n = len(words)
    ans = float('inf')
    for i in range(n):
        if words[i]==target:
            sr = abs(i-startIndex)
            _min = min(sr, n-sr)
            if ans > _min:
                ans = _min
    return -1 if ans == float('inf') else ans
```

```python
words = ["hello","i","am","leetcode","hello"]
target = "hello"
startIndex = 1
closestTarget(words, target, startIndex)
```

```python
words = ["i","eat","leetcode"]
target = "ate"
startIndex = 0
closestTarget(words, target, startIndex)
```
