```python
def countStudents(students, sandwiches):
    n = len(students)
    student0, student1 = 0, 0
    for s in students:
        if s == 0:
            student0 += 1
        else:
            student1 += 1
            
    for s in sandwiches:
        if s==0:
            if student0>0:
                student0 -= 1
            else:
                break
        else:
            if student1>0:
                student1 -= 1
            else:
                break
            
    return student0 + student1
```

```python
students = [1,1,1,0,0,1]
sandwiches = [1,0,0,0,1,1]
countStudents(students, sandwiches)
```

```python
students = [1,1,0,0]
sandwiches = [0,1,0,1]
countStudents(students, sandwiches)
```
