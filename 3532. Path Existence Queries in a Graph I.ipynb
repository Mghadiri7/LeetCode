```python
def pathExistenceQueries(n, nums, maxDiff, queries):
    component_ids = [0] * n
    current_id = 0

    for i in range(n-1):
        if nums[i+1] - nums[i] > maxDiff:
            current_id += 1
        component_ids[i+1] = current_id

    res = []
    for u, v in queries:
        res.append(component_ids[u]==component_ids[v])

    return res
```

```python
n = 4
nums = [2,5,6,8]
maxDiff = 2
queries = [[0,1],[0,2],[1,3],[2,3]]
pathExistenceQueries(n, nums, maxDiff, queries)
```
