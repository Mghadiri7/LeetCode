```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def modifiedList(self, nums, head):
        _set = set(nums)
        _linklist = ListNode(0)
        _linklist.next = head
        temp = _linklist
        while temp.next:
            if temp.next.val in _set:
                temp.next = temp.next.next
            else:
                temp = temp.next
        return _linklist.next
```

```python
def build_linked_list(values):
    dummy = ListNode(0)
    current = dummy
    for v in values:
        current.next = ListNode(v)
        current = current.next
    return dummy.next

def linked_list_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result


nums = [1, 2, 3]
head_values = [1, 2, 3, 4, 5]

head = build_linked_list(head_values)
sol = Solution()
new_head = sol.modifiedList(nums, head)

print(linked_list_to_list(new_head))
```
