```python
def bestClosingTime(customers):
    min_penalty = 0
    min_time = 0
    prefix = 0
    for i, c in enumerate(customers):
        if c == 'Y':
            prefix -= 1
        else:
            prefix += 1
            
        if prefix < min_penalty:
            min_time = i+1
            min_penalty = prefix
            
    return min_time
```

```python
customers = "YYNY"
bestClosingTime(customers)
```
