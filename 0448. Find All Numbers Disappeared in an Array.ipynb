```python
def findDisappearedNumbers(nums):
    n = len(nums)
    lst = [0] * (n+1)
    ans = []
    for num in nums:
        lst[num] += 1
    for i in range(1, n+1):
        if lst[i]==0:
            ans.append(i)
    return ans
```

```python
nums = [4,3,2,7,8,2,3,1]
findDisappearedNumbers(nums)
```
