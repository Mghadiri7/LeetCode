## First Solution

```python
def isUgly(n):
    if n<1:
        return False
    if n==1:
        return True
    elif n%2==0:
        return isUgly(n//2)
    elif n%3==0:
        return isUgly(n//3)
    elif n%5==0:
        return isUgly(n//5)
    else:
        return False
```

```python
isUgly(6)
```

## Second Solution

```python
def isUgly(n):
    if n<1:
        return False
    while n%2==0:
        n//=2
    while n%3==0:
        n//=3
    while n%5==0:
        n//=5
    return n==1
```

```python
isUgly(14)
```
