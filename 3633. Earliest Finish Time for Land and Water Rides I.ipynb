```python
def earliestFinishTime(landStartTime, landDuration, waterStartTime, waterDuration):
    min_land, min_water, res = 3000, 3000, 3000
    n, m = len(landStartTime), len(waterStartTime)
    
    for i in range(n):
        end_land = landStartTime[i] + landDuration[i]
        if end_land < min_land:
            min_land = end_land
            
    for j in range(m):
        end_water = waterStartTime[j] + waterDuration[j]
        if end_water < min_water:
            min_water = end_water

    for j in range(m):
        end_lw = max(min_land, waterStartTime[j]) + waterDuration[j]
        if end_lw < res:
            res = end_lw

    for i in range(n):
        end_wl = max(min_water, landStartTime[i]) + landDuration[i]
        if end_wl < res:
            res = end_wl

    return res
```

```python
landStartTime = [5]
landDuration = [3]
waterStartTime = [1]
waterDuration = [10]
earliestFinishTime(landStartTime, landDuration, waterStartTime, waterDuration)
```
