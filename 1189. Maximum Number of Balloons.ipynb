## First Solution

```python
def maxNumberOfBalloons(text):
    lower_text = text.lower()
    _dict = {"b":0, "a":0, "l":0, "o":0, "n":0}
    for l in lower_text:
        if l == "b":
            _dict["b"]+=1
        elif l == "a":
            _dict["a"]+=1
        elif l == "l":
            _dict["l"]+=1
        elif l == "o":
            _dict["o"]+=1
        elif l == "n":
            _dict["n"]+=1
    _dict["l"] //= 2
    _dict["o"] //= 2

    ans = _dict["b"]
    for value in _dict.values():
        if value < ans:
            ans = value
    return ans
```

```python
text = "loonbalxballpoon"
maxNumberOfBalloons(text)
```

## Second Solution

```python
from collections import Counter

def maxNumberOfBalloons(text):
    _dict = Counter(text)
    return min(_dict[k]//v for k,v in {"b":1, "a":1, "l":2, "o":2, "n":1}.items())
```

```python
text = "loonbalxballpoon"
maxNumberOfBalloons(text)
```
