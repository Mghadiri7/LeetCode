```python
def leftRightDifference(nums):
    n = len(nums)
    lst = [0] * n
    left_sum = 0
    right_sum = sum(nums)
    for i in range(n):
        diff = nums[i]
        right_sum -= diff
        lst[i] = abs(right_sum - left_sum)
        left_sum += diff
    return lst
```

```python
nums = [10,4,8,3]
leftRightDifference(nums)
```
