```python
from heapq import *

def isPossible(target):
    if len(target) == 1:
        return target[0] == 1
        
    total = sum(target)
    hp = [-x for x in target]
    heapify(hp)
    
    while -hp[0]>1:
        max_val = -heappop(hp)
        rest = total - max_val 
        if rest==0 or max_val<=rest:
            return False
        if rest==1:
            return True
            
        new_val = max_val % rest
        if new_val==0:
            return False
        total = total - max_val + new_val
        heappush(hp, -new_val)
        
    return True
```

```python
target = [9,3,5]
isPossible(target)
```
