```python
def findMaxForm(strs, m, n):
    dp = {(0, 0): 0}
    for s in strs:
        zeroes = ones = 0
        new_dp = {}
        for ch in s:
            if ch=="0":
                zeroes+=1
            else:
                ones+=1
        for (pre_zeroes, pre_ones), val in dp.items():
            new_zeroes, new_ones = pre_zeroes+zeroes, pre_ones+ones
            if new_zeroes <= m and new_ones <= n:
                if (new_zeroes, new_ones) not in dp or dp[(new_zeroes, new_ones)] < val+1:
                    new_dp[(new_zeroes, new_ones)] = val+1
        dp.update(new_dp)
    return max(dp.values())
```

```python
strs = ["10","0001","111001","1","0"]
m = 5
n = 3
findMaxForm(strs, m, n)
```
