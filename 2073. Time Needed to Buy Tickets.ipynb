```python
def timeRequiredToBuy(tickets, k):
    ans = 0
    _max = tickets[k]
    for i, t in enumerate(tickets):
        if i<=k:
            ans += min(tickets[i], _max)
        else:
            ans += min(tickets[i], _max - 1)
    return ans
```

```python
tickets = [5,1,1,1]
k = 0
timeRequiredToBuy(tickets, k)
```
