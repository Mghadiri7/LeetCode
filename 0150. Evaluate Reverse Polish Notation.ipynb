```python
def evalRPN(tokens):
    stack = []
    lst = ["+", "-", "*", "/"]
    for x in tokens:
        if x in lst:
            num1 = stack.pop()
            num2 = stack.pop()
            if x=="+":
                stack.append(num2 + num1)
            elif x=="-":
                stack.append(num2 - num1)
            elif x=="*":
                stack.append(num2 * num1)
            else:
                stack.append(int(num2 / num1))
        else:
            stack.append(int(x))
    return stack[0]
```

```python
tokens = ["2","1","+","3","*"]
evalRPN(tokens)
```

```python
tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
evalRPN(tokens)
```
