## First Solution

```python
def closestPrimes(left, right):
    def is_prime(n):
        if n <= 1:
            return False
        if n <= 3:
            return True
        if n % 2 == 0 or n % 3 == 0:
            return False
    
        i = 5
        while i * i <= n: 
            if n % i == 0:
                return False
            if n % (i + 2) == 0:
                return False
            i += 6
        return True

    primes = []
    for num in range(max(2, left), right + 1):
        if is_prime(num):
            primes.append(num)

    if len(primes) < 2:
        return [-1, -1]

    _min = 1000001
    ans = [-1, -1]
    for i in range(1, len(primes)):
        gap = primes[i] - primes[i-1]
        if gap < _min:
            _min = gap
            result = [primes[i-1], primes[i]]
        
    return result
```

```python
left = 10
right = 19
closestPrimes(left, right)
```

## Second Solution

```python
def closestPrimes(left, right):
    if right<=2:
        return [-1,-1]

    right += 1
    seen = [1] * (right)
    seen[0] = seen[1] = 0
    for i in range(2, right):
        if seen[i]==1:
            for j in range(i*2, right, i):
                seen[j] = 0

    prev = -1
    _min = 1000001
    ans = [-1, -1]

    for i in range(left, right):
        if seen[i] == 1:
            if prev != -1:
                dist = i - prev
                if dist < _min:
                    _min = dist
                    ans = [prev, i]
            prev = i
    return ans
```

```python
left = 10
right = 19
closestPrimes(left, right)
```
