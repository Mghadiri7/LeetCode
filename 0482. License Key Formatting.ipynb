```python
def licenseKeyFormatting(s, k):
    s = s.replace('-', '')
    s = s.upper()
    n = len(s)
    head = n % k
    lst = []
    if head:
        lst.append(s[:head])
    for i in range(head, n, k):
        lst.append(s[i:i+k])

    return '-'.join(lst)
```

```python
s = "5F3Z-2e-9-w"
k = 4
licenseKeyFormatting(s, k)
```
