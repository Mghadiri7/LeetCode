```python
def merge(nums1, m, nums2, n):
    length = m + n - 1
    while m>0 and n>0:
        if nums1[m-1] > nums2[n-1]:
            nums1[m-1], nums1[length] = nums1[length], nums1[m-1]
            m-=1
        else:
            nums1[length] = nums2[n-1]
            n-=1
        length-=1
        
    if m==0 and n>0:
        # for i in range(n):
        #     nums1[i] = nums2[i]
        nums1[:n] = nums2[:n]
            
    print(nums1)
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3
merge(nums1, m, nums2, n)
```
