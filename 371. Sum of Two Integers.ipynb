```python
def getSum(a, b):
    MASK = 0xFFFFFFFF 
    MAX_INT = 0x7FFFFFFF 

    while b != 0:
        carry = (a & b) << 1
        a = a ^ b
        b = carry
        a = a & MASK
        b = b & MASK

    if a > MAX_INT:
        return ~(a ^ MASK)
    else:
        return a
```

```python
a = 2
b = 3
getSum(a, b)
```
