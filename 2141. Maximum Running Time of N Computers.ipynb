```python
def maxRunTime(n, batteries):
    batteries.sort(reverse=True)
    total = sum(batteries)
    for b in batteries:
        if b <= total//n:
            break
        n -= 1
        total -= b
        
    return total//n
```

```python
n = 2
batteries = [3,3,3]
maxRunTime(n, batteries)
```

```python
def maxRunTime(n, arr):
    arr.sort()
    total = sum(arr)

    while arr[-1] > total // n:
        n -= 1
        total -= arr.pop()

    return total // n
```
