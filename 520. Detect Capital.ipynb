```python
def detectCapitalUse(word):
    counter = 0
    for s in word:
        if s.isupper():
            counter+=1
    if counter==len(word) or counter==0:
        return True
    elif counter==1 and word[0].isupper():
        return True
    else:
        return False
```

```python
word = 'fffffF'
detectCapitalUse(word)
```
