## First Solution

```python
def countMajoritySubarrays(nums, target):
    n = len(nums)
    ans = 0

    for l in range(n):
        count_target = 0
        for r in range(l, n):
            if nums[r] == target:
                count_target+=1
            length = r-l+1
            if count_target > length//2:
                ans+=1
    
    return ans
```

```python
nums = [1,2,2,3]
target = 2
countMajoritySubarrays(nums, target)
```

## Second Solution

```python
def countMajoritySubarrays(nums, target):
    n = len(nums)
    ans = 0
    lst = [0] * (n+1)

    for i in range(n):
        lst[i+1] = lst[i] + (1 if nums[i]==target else -1)

    for l in range(n):
        for r in range(l, n):
            if lst[r+1]>lst[l]:
                ans+=1
    return ans
```

```python
nums = [1,2,2,3]
target = 2
countMajoritySubarrays(nums, target)
```

## Third Solution

```python
def countMajoritySubarrays(nums, target):
    n = len(nums)
    ans = 0
    lst = [0] * (2*n+1)
    cnt = n
    lst[cnt] = 1
    presum = 0

    for x in nums:
        if x == target:
            presum += lst[cnt]
            cnt+=1
            lst[cnt] += 1

        else:
            cnt-=1
            presum -= lst[cnt]
            lst[cnt]+=1

        ans += presum
    return ans
```

```python
nums = [2,2,2,2]
target = 2
countMajoritySubarrays(nums, target)
```
