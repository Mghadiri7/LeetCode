```python
def maxFreqSum(s):
    dict_vowel = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}
    dict_consonant = {}
    max_vowel = 0
    max_consonant = 0
    for i in s:
        if i in dict_vowel:
            dict_vowel[i]+=1
        elif i in dict_consonant:
            dict_consonant[i]+=1
        else:
            dict_consonant[i]=1
    for i in dict_vowel:
        if max_vowel<dict_vowel[i]:
            max_vowel=dict_vowel[i]
    for i in dict_consonant:
        if max_consonant<dict_consonant[i]:
            max_consonant=dict_consonant[i]
    return max_vowel+max_consonant
```

```python
s = "successes"
maxFreqSum(s)
```
