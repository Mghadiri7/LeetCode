```python
def minDeletionSize(strs):
    res = 0
    cols = len(strs[0])
    rows = len(strs)
    sorted_pairs = [False] * (rows - 1)
    for col in range(cols):
        flag = False
        for i in range(rows-1):
            if not sorted_pairs[i] and strs[i][col] > strs[i+1][col]:
                flag = True
                break
        if flag:
            res += 1
            continue
        for i in range(rows-1):
            if not sorted_pairs[i] and strs[i][col] < strs[i+1][col]:
                sorted_pairs[i] = True
        if all(sorted_pairs):
            break
    return res
```

```python
strs = ["ca","bb","ac"]
minDeletionSize(strs)
```
