```python
def minDeletionSize(strs):
    res = 0
    cols = len(strs[0])
    rows = len(strs)
    for col in range(cols):
        for row in range(1, rows):
            if strs[row][col] < strs[row-1][col]:
                res += 1
                break
    return res
```

```python
strs = ["cba","daf","ghi"]
minDeletionSize(strs)
```
