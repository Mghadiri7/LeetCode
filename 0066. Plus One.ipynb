## First solution

```python
def plusOne(digits):
    n = len(digits)-1
    lst = []
    for i in range(n, -1, -1):
        if digits[i]<9:
            digits[i]+=1
            lst[0:0] = digits[0:i+1]
            break
        else:
            lst.append(0)
            if i==0:
                lst.insert(0, 1)
    return lst
```

```python
digits = [1,2,3]
plusOne(digits)
```

```python
digits = [1,2,3,9,9,9]
plusOne(digits)
```

```python
digits = [9,9,9]
plusOne(digits)
```

## Second Solution

```python
def plusOne(digits):
    n = len(digits)-1
    for i in range(n, -1, -1):
        if digits[i]<9:
            digits[i]+=1
            return digits
        
        digits[i] = 0
        if i==0:
            return [1]+digits
```

```python
digits = [9,9,9]
plusOne(digits)
```

```python
digits = [1,2,3,9,9,9]
plusOne(digits)
```
