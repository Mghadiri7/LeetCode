```python
def getDescentPeriods(prices):
    ans = 0
    counter = 1

    for i in range(len(prices)):
        if i == 0:
            ans += counter
            continue

        if prices[i] == prices[i-1]-1:
            counter+=1
        else:
            counter = 1
        ans += counter
    return ans
```

```python
prices = [3,2,1,4]
getDescentPeriods(prices)
```
