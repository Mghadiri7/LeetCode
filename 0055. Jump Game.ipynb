```python
def canJump(nums):
    n = len(nums)
    farthest = 0
    for i in range(n):
        if i>farthest:
            return False
        elif farthest>=n-1:
            return True
        else:
            farthest = max(farthest, i+nums[i])
```

```python
nums = [2,3,1,1,4]
canJump(nums)
```

```python
nums = [3,2,1,0,4]
canJump(nums)
```
