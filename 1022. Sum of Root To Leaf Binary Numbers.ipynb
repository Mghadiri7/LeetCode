```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def sumRootToLeaf(self, root):
        def dfs(node, curr_sum):
            if not node:
                return 0
    
            curr_sum = curr_sum * 2 + node.val
            if not node.left and not node.right:
                return curr_sum

            return dfs(node.left, curr_sum) + dfs(node.right, curr_sum)

        return dfs(root, 0)
```

```python
from collections import deque

def _is_null(x):
    return x is None or x == 'null'

def build_tree(values):
    if not values or _is_null(values[0]):
        return None

    root = TreeNode(values[0])
    q = deque([root])
    i = 1

    while q and i < len(values):
        node = q.popleft()

        if i < len(values) and not _is_null(values[i]):
            node.left = TreeNode(values[i])
            q.append(node.left)
        i += 1

        if i < len(values) and not _is_null(values[i]):
            node.right = TreeNode(values[i])
            q.append(node.right)
        i += 1

    return root
```

```python
values = [1,0,1,0,1,0,1]
root = build_tree(values)
sol = Solution()
print(sol.sumRootToLeaf(root))
```
