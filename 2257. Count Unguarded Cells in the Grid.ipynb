```python
def countUnguarded(m, n, guards, walls):
    lst = [[0] * n for i in range(m)]
    for row, col in guards+walls:
        lst[row][col] = 2

    directions = [(1,0), (-1,0), (0,1), (0,-1)]
    for row, col in guards:
        for dr, dc in directions:
            new_row = row+dr
            new_col = col+dc
            while 0<=new_row<m and 0<=new_col<n and lst[new_row][new_col]!=2:
                lst[new_row][new_col]=1
                new_row+=dr
                new_col+=dc
    return sum(row.count(0) for row in lst)
    # print(lst)
    # res = 0
    # for i in range(m):
    #     for j in range(n):
    #         if lst[i][j]==0:
    #             res+=1
    # return res
```

```python
m = 4
n = 6
guards = [[0,0],[1,1],[2,3]]
walls = [[0,1],[2,2],[1,4]]
countUnguarded(m, n, guards, walls)
```
