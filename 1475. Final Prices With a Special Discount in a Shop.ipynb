```python
def finalPrices(prices):
    n = len(prices)
    answer = [0] * n
    for i in range(n):
        discount = 0
        for j in range(i+1, n):
            if prices[i] >= prices[j]:
                discount = prices[j]
                break
        answer[i] = prices[i] - discount
    return answer
```

```python
prices = [8,4,6,2,3]
finalPrices(prices)
```

```python
prices = [8,7,4,2,8,1,7,7,10,1]
finalPrices(prices)
```
