```python
def hasSameDigits(s):
    lst = [int(digit) for digit in s]
    n = len(lst)-1
    for i in range(n, 1, -1):
        for j in range(i):
            lst[j] = (lst[j] + lst[j+1]) % 10
    if lst[0]==lst[1]:
        return True
    return False
```

```python
s = "3902"
hasSameDigits(s)
```
