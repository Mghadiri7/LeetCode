```python
def intersectionSizeTwo(intervals):
    intervals.sort(key= lambda x: (x[1], -x[0]))

    ans = 0
    l, r = -1, -1
    for left, right in intervals:
        if left>r:
            l = right-1
            r = right
            ans+=2
        if left>l:
            l = r
            r = right
            ans+=1
    return ans
```

```python
intervals = [[1,3],[1,4],[2,5],[3,5]]
intersectionSizeTwo(intervals)
```

```python
intervals = [[1,2],[2,3],[2,4],[4,5]]
intersectionSizeTwo(intervals)
```
