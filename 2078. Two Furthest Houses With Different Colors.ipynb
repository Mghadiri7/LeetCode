```python
def maxDistance(colors):
    n = len(colors)
    first = colors[0]
    last = colors[-1]
    max1, max2 = 0, 0
    for i in range(n-1, 0, -1):
        if first != colors[i]:
            max1 = i
            break
    for i in range(n):
        if last != colors[i]:
            max2 = n-i-1
            break
    if max1>max2:
        return max1
    return max2
```

```python
colors = [4,4,4,11,4,4,11,4,4,4,4,4]
maxDistance(colors)
```
