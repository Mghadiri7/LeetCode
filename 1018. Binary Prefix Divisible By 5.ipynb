## First Solution

```python
def prefixesDivBy5(nums):
    n = len(nums)
    res = 0
    ans = []
    for x in nums:
        res = res * 2 + x
        if res % 5 == 0:
            ans.append(True)
        else:
            ans.append(False)
    return ans
```

```python
nums = [0,1,1]
prefixesDivBy5(nums)
```

```python
nums = [1,1,0,0,0,1,0,0,1]
prefixesDivBy5(nums)
```

## Second Solution

```python
def prefixesDivBy5(nums):
    n = len(nums)
    ans = [0] * n
    res = 0
    for i in range(n):
        res = res * 2 + nums[i]
        ans[i] = (res % 5 == 0)
    return ans
```

```python
nums = [0,1,1]
prefixesDivBy5(nums)
```

```python
nums = [1,1,0,0,0,1,0,0,1]
prefixesDivBy5(nums)
```
