```python
def maxOperations(s):
    ones = ans = 0
    flag = False
    for ch in s:
        if ch=='1':
            ones+=1
            flag = True
        elif flag:
            ans += ones
            flag = False
    return ans
```

```python
s = "1001101"
maxOperations(s)
```

```python
s = "00111"
maxOperations(s)
```
