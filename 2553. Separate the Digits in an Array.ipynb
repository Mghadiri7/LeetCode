## First Solution

```python
def separateDigits(nums):
    res = []
    for x in nums:
        lst = []
        while x>0:
            lst.append(x%10)
            x //= 10
        lst.reverse()
        res.extend(lst)
    return res
```

```python
nums = [13,25,83,77]
separateDigits(nums)
```

## Second Solution

```python
def separateDigits(nums):
    res = []
    for x in nums:
        res.extend(int(digit) for digit in str(x))
        
    return res
```

```python
nums = [13,25,83,77]
separateDigits(nums)
```
