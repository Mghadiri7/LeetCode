```python
import math

def maxSubarraySum(nums, k):
    current_prefix_sum = 0
    max_sum = -math.inf
    min_prefixes = [math.inf] * k
    min_prefixes[0] = 0
    
    for i, num in enumerate(nums):
        current_prefix_sum += num
        remainder = (i + 1) % k
        if min_prefixes[remainder] != math.inf:
            candidate = current_prefix_sum - min_prefixes[remainder]
            max_sum = max(max_sum, candidate)
        
        min_prefixes[remainder] = min(min_prefixes[remainder], current_prefix_sum)
            
    return max_sum if max_sum != -math.inf else 0
```

```python
nums = [1, -2, 4, 3, -1]
k = 2
maxSubarraySum(nums, k)
```
