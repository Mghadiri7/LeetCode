```python
def buildArray(target, n):
    n+=1
    length = len(target)
    res = []
    j = 0
    for i in range(1, n):
        if target[j]==i:
            res.append("Push")
            j+=1
            if j==length:
                break
        else:
            res.append("Push")
            res.append("Pop")
    return res
```

```python
target = [1,3]
n = 3
buildArray(target, n)
```

```python
target = [1,2]
n = 4
buildArray(target, n)
```
