```python
def minCost(colors, neededTime):
    n = len(colors)-1
    res = 0
    i = 0
    while i<n:
        j = i+1
        _sum = neededTime[i]
        _max = neededTime[i]
        while j<=n and colors[i]==colors[j]:
            _sum += neededTime[j]
            if neededTime[j]>_max:
                _max = neededTime[j]
            j+=1
        res += _sum - _max
        i = j
    return res
```

```python
colors = "aabaa"
neededTime = [1,2,3,4,1]
minCost(colors, neededTime)
```

```python
colors = "abaac"
neededTime = [1,2,3,4,5]
minCost(colors, neededTime)
```
