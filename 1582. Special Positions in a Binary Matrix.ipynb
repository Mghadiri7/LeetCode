```python
def numSpecial(mat):
    res = 0
    m = len(mat)
    n = len(mat[0])
    row_special = [0] * m
    lst = [[] for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if mat[i][j] == 1:
                lst[i].append(j)
        
        if sum(mat[i]) == 1:
            row_special[i] = 1

    col_special = [0] * n
    for row in lst:
        for col in row:
            col_special[col] += 1
    
    for i in range(m):
        if row_special[i] == 1:
            check = lst[i][0]
            if col_special[check] == 1:
                res += 1
                
    return res
```

```python
mat = [[1,0,0],[0,1,0],[0,0,1]]
numSpecial(mat)
```
