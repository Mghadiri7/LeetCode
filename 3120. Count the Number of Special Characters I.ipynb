## First solution

```python
def numberOfSpecialChars(word):
    ans = 0
    for i in range(65, 97):
        if chr(i) in word and chr(i+32) in word:
            ans += 1
    return ans
```

```python
word = "aaAbcBC"
numberOfSpecialChars(word)
```

## Second solution

```python
import string

def numberOfSpecialChars(word):
    unique_chars = set(word)
    res = 0
    for char in string.ascii_lowercase:
        if char in unique_chars and char.upper() in unique_chars:
            res +=1
    return res
```

```python
word = "abBCab"
numberOfSpecialChars(word)
```
