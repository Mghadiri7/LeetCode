```python
def isOneBitCharacter(bits):
    n = len(bits)-1
    i = 0
    while i<n:
        if bits[i]==0:
            i+=1
        if bits[i]==1:
            i+=2
    if i==n:
        return True
    return False
```

```python
bits = [1,1,1,0]
isOneBitCharacter(bits)
```

```python
bits = [1,0,0]
isOneBitCharacter(bits)
```
