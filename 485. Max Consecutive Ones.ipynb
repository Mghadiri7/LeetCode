```python
def findMaxConsecutiveOnes(nums):
    ans = 0
    counter = 0
    for x in nums:
        if x == 0:
            counter = 0
        else:
            counter+=1
            if counter>ans:
                ans = counter
    return ans
```

```python
nums = [1,1,0,1,1,1]
findMaxConsecutiveOnes(nums)
```
