```python
def mapWordWeights(words, weights):
    lst = []
    for word in words:
        total_weight = 0
        for char in word:
            index = ord(char) - ord('a')
            total_weight += weights[index]

        mod = total_weight % 26
        map_char = chr(ord('z')-mod)
        lst.append(map_char)

    return "".join(lst)
```

```python
words = ["abcd","def","xyz"]
weights = [5,3,12,14,1,2,3,2,10,6,6,9,7,8,7,10,8,9,6,9,9,8,3,7,7,2]
mapWordWeights(words, weights)
```
