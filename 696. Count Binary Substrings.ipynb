## First Solution

```python
def countBinarySubstrings(s):
    lst = []
    n = len(s)
    ones = 0
    zeroes = 0
    if s[0]=='0':
        zeroes = 1
    else:
        ones = 1

    for i in range(1, n):
        if s[i]==s[i-1]:
            if s[i]=='0':
                zeroes+=1
            else:
                ones+=1
        else:
            if s[i]=='0':
                lst.append(ones)
                ones = 0
                zeroes = 1
            else:
                lst.append(zeroes)
                zeroes = 0
                ones = 1
    if zeroes > 0:
        lst.append(zeroes)
    else:
        lst.append(ones)

    res = 0
    m = len(lst) - 1
    for i in range(m):
        res += min(lst[i], lst[i+1])
    return res
```

```python
s = "00110011"
countBinarySubstrings(s)
```

## Second Solution

```python
def countBinarySubstrings(s):
    res = 0
    prev = 0
    curr = 1
    n = len(s)
    for i in range(1, n):
        if s[i]==s[i-1]:
            curr += 1
        else:
            res += min(prev, curr)
            prev = curr
            curr = 1
    res += min(prev, curr)
    return res
```

```python
s = "00110011"
countBinarySubstrings(s)
```
