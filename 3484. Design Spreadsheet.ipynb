```python
class Spreadsheet:
    def __init__(self, rows: int):
        self.dict = {}
    def setCell(self, cell: str, value: int) -> None:
        self.dict[cell] = value
    def resetCell(self, cell: str) -> None:
        self.dict[cell] = 0
    def getValue(self, formula: str) -> int:
        ops = formula.split('=')[1].split('+')
        res = 0
        for op in ops:
            if op in self.dict:
                res += self.dict[op]
            elif op.isdigit():
                res += int(op)
        return res
```

```python
cmds = ["Spreadsheet","getValue","setCell","getValue","setCell","getValue","resetCell","getValue"]
args = [[3],["=5+7"],["A1",10],["=A1+6"],["B2",15],["=A1+B2"],["A1"],["=A1+B2"]]
out = []
obj = None
for cmd, arg in zip(cmds, args):
    if cmd == "Spreadsheet":
        obj = Spreadsheet(*arg)
        out.append(None)
    elif cmd == "setCell":
        out.append(obj.setCell(*arg))
    elif cmd == "resetCell":
        out.append(obj.resetCell(*arg))
    elif cmd == "getValue":
        out.append(obj.getValue(*arg))

print(out)
```
