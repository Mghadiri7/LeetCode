```python
def sumAndMultiply(s, queries):
    MOD = 10**9 + 7
    n = len(s)
    s_prime = []
    prefix_count = [0] * (n + 1)
    
    count = 0
    for i, char in enumerate(s):
        if char != '0':
            s_prime.append(int(char))
            count += 1
        prefix_count[i + 1] = count

    K = len(s_prime)
    pow10 = [1] * (K + 1)
    for i in range(1, K + 1):
        pow10[i] = (pow10[i - 1] * 10) % MOD

    pref_val = [0] * (K + 1)
    pref_sum = [0] * (K + 1)

    for i in range(K):
        pref_val[i + 1] = (pref_val[i] * 10 + s_prime[i]) % MOD
        pref_sum[i + 1] = pref_sum[i] + s_prime[i]

    ans = []
    for l, r in queries:
        L = prefix_count[l]
        R = prefix_count[r + 1] - 1

        if L > R:
            ans.append(0)
        else:
            length = R - L + 1
            x = (pref_val[R + 1] - (pref_val[L] * pow10[length]) % MOD) % MOD
            x = (x + MOD) % MOD
            sum_digits = pref_sum[R + 1] - pref_sum[L]
            ans.append((x * sum_digits) % MOD)

    return ans
```

```python
s = "10203004"
queries = [[0,7],[1,3],[4,6]]
sumAndMultiply(s, queries)
```
