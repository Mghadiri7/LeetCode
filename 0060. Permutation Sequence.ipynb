```python
from math import factorial

def getPermutation(n, k):
    k -= 1
    lst = [1,2,3,4,5,6,7,8,9]
    ans = ""
    f = factorial(n-1)
    for i in range(n):
        num = k//f
        ans += str(lst[num])
        del lst[num]
        k %= f
        if i<n-1:
            f //= (n-i-1)
    return ans
```

```python
n = 4
k = 9
getPermutation(n, k)
```
