## First Solution

```python
from functools import lru_cache

def rearrangeSticks(n, k):
    mod = 10**9+7
    @lru_cache(maxsize=None)
    def dp(num_stick, visible):
        if num_stick==visible:
            return 1
        if visible==0:
            return 0
        else:
            return (dp(num_stick-1, visible-1) + dp(num_stick-1, visible) * (num_stick-1)) % mod
    return dp(n, k) % mod
```

```python
n = 20
k = 11
rearrangeSticks(n, k)
```

## Second Solution

```python
def rearrangeSticks(n, k):
    mod = 10**9+7
    lst = [[0]*(k+1) for i in range(n+1)]
    lst[0][0] = 1
    for i in range(1, n+1):
        for j in range(1, k+1):
            lst[i][j] = (lst[i-1][j-1] + lst[i-1][j] * (i-1)) % mod
    return lst[n][k]
```

```python
n = 20
k = 11
rearrangeSticks(n, k)
```
