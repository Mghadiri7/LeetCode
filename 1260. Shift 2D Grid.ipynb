```python
def shiftGrid(grid, k):
    m = len(grid)
    n = len(grid[0])
    lst = [[0] * n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            lst[(i + ((j+k) // n)) % m][(j+k) % n] = grid[i][j]
    return lst
```

```python
grid = [[1,2,3],[4,5,6],[7,8,9]]
k = 1
shiftGrid(grid, k)
```
