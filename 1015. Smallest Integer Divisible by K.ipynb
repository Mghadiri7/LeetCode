```python
def smallestRepunitDivByK(k):
    if k%10 not in {1,3,7,9}:
        return -1

    _mod, mod_set = 0, set()
    for i in range(1, k+1):
        _mod = (10 * _mod + 1) % k
        if _mod==0:
            return i
        if _mod in mod_set:
            return -1
        mod_set.add(_mod)
    return -1
```

```python
k = 3
smallestRepunitDivByK(k)
```
