## First solution

```python
from math import gcd

def minOperations(nums):
    n = len(nums)
    ones = nums.count(1)
    if ones:
        return n-ones

    res = float('inf')
    for i in range(n):
        g = nums[i]
        for j in range(i+1, n):
            g = gcd(g, nums[j])
            if g==1:
                res = min(res, j-i)

    if res==float('inf'):
        return -1

    return res+n-1
```

```python
nums = [4,6,9]
minOperations(nums)
```

## Second Solution

```python
from math import gcd

def minOperations(nums):
    n = len(nums)
    ones = 0
    g = 0
    for x in nums:
        if x==1:
            ones+=1
        g = gcd(g, x)

    if ones>0:
        return n-ones
    if g>1:
        return -1

    res = n
    for i in range(n):
        g = nums[i]
        for j in range(i+1, n):
            g = gcd(g, nums[j])
            if g==1:
                res = min(res, j-i)
                break

    return res+n-1
```

```python
nums = [4,6,9]
minOperations(nums)
```
