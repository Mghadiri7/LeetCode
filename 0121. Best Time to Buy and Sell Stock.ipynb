```python
def maxProfit(prices):
    res = 0
    n = len(prices)
    current = prices[0]
    for i in range(1, n):
        if current > prices[i]:
            current = prices[i]
        else:
            res = max(prices[i] - current, res)
            
    return res
```

```python
prices = [7,1,5,3,6,4]
maxProfit(prices)
```
