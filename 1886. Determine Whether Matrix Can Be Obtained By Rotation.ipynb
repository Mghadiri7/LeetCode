```python
def findRotation(mat, target):
    if mat==target:
        return True
        
    n = len(mat)
    for k in range(3):
        for i in range(n):
            for j in range(i+1, n):
                mat[i][j], mat[j][i] = mat[j][i], mat[i][j]
                
        for i in range(n):
            # matrix[i] = matrix[i][::-1]
            mat[i].reverse()

        if mat==target:
            return True

    return False
```

```python
mat = [[0,0,0],[0,1,0],[1,1,1]]
target = [[1,1,1],[0,1,0],[0,0,0]]
findRotation(mat, target)
```
