## Frist Solution

```python
def countPalindromicSubsequence(s):
    n = len(s)
    first = [n] * 26
    last = [-1] * 26
    for i, ch in enumerate(s):
        idx = ord(ch) - ord('a')
        first[idx] = min(first[idx], i)
        last[idx] = i

    ans = 0
    for i in range(26):
        left, right = first[i], last[i]
        if right-left<2:
            continue

        seen = [False] * 26
        for j in range(left+1, right):
            idx = ord(s[j]) - ord('a')
            if not seen[idx]:
                seen[idx] = True
                ans+=1

    return ans
```

```python
s = "bbcbaba"
countPalindromicSubsequence(s)
```

## Second Solution

```python
def countPalindromicSubsequence(s):
    ans = 0
    chars = set(s)
    for ch in chars:
        left = s.find(ch)
        right = s.rfind(ch)
        ans += len(set(s[left+1:right]))
    return ans    
```

```python
s = "bbcbaba"
countPalindromicSubsequence(s)
```
