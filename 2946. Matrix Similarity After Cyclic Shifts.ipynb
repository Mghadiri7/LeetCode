```python
def areSimilar(mat, k):
    m = len(mat[0])
    k = k % m
    if k==0:
        return True

    n = len(mat)
    lst = mat.copy()
    for i in range(0, n, 2):
        lst[i] = lst[i][k:] + lst[i][:k]
    for j in range(1, n, 2):
        lst[j] = lst[j][-k:] + lst[j][:-k]
    for i in range(n):
        if lst[i] != mat[i]:
            return False
            
    return True
```

```python
mat = [[1,2,1,2],[5,5,5,5],[6,3,6,3]]
k = 2
areSimilar(mat, k)
```
