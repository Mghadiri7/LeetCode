```python
def projectionArea(grid):
    n = len(grid)
    area = 0
    for row in grid:
        area += max(row)
        for num in row:
            if num>0:
                area += 1
    for col in range(n):
        column = [row[col] for row in grid]
        area += max(column)
    return area
```

```python
grid = [[1,2],[3,4]]
projectionArea(grid)
```
