```python
def getMinDistance(nums, target, start):
    n = len(nums)
    ans = float('inf')
    for i in range(n):
        if nums[i]==target:
            if abs(i - start) < ans:
                ans = abs(i - start)
    return ans
```

```python
nums = [1,2,3,4,5]
target = 5
start = 3
getMinDistance(nums, target, start)
```

```python
nums = [5,7,7,5]
target = 5
start = 2
getMinDistance(nums, target, start)
```
