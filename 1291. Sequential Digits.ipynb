```python
def sequentialDigits(low, high):
    l = len(str(low))
    h = len(str(high)) + 1
    res = []
    digit = "123456789"
    for length in range(l, h):
        for start in range(10-length):
            num = int(digit[start : start+length])
            if low <= num <= high:
                res.append(num)

    return res
```

```python
low = 1000
high = 13000
sequentialDigits(low, high)
```
