## First Solution

```python
def isGood(nums):
    n = len(nums)
    lst = [0] * n
    for x in nums:
        if x < n:
            lst[x] += 1
    print(lst)

    if lst[-1] != 2:
        return False

    for i in range(1, n-1):
        if lst[i] != 1:
            return False

    return True
```

```python
nums = [3, 4, 4, 1, 2, 1]
isGood(nums)
```

```python
nums = [1, 3, 3, 2]
isGood(nums)
```

## Second Solution

```python
def isGood(nums):
    n = len(nums) - 1
    for x in nums:
        if x < 1 or x > n:
            return False

    for i in range(n+1):
        val = abs(nums[i])
        if nums[val-1] < 0:
            if val != n:
                return False

        nums[val-1] *= -1

    for i in range(n - 1):
        if nums[i] > 0:
            return False

    return nums[n-1] > 0
```

```python
nums = [3, 4, 4, 1, 2, 1]
isGood(nums)
```
