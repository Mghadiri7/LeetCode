## First Solution

```python
def sumAndMultiply(n):
    x = ""
    _sum = 0
    for s in str(n):
        if s!='0':
            x += s
    if x=="":
        x = '0'
    for s in x:
        _sum += int(s)
    return int(x) * _sum
```

```python
n = 10203040
sumAndMultiply(n)
```

## Second Solution

```python
def sumAndMultiply(n):
    x = 0
    _sum = 0
    place = 1

    while n>0:
        d = n % 10
        if d != 0:
            x = d * place + x
            place *= 10
            _sum += d

        n //= 10

    return x * _sum
```

```python
n = 10203040
sumAndMultiply(n)
```
