```python
def minTime(skill, mana):
    n = len(skill)
    m = len(mana)
    sumSkill = sum(skill)
    prev = sumSkill * mana[0]
    for j in range(1, m):
        prevPotionDone = prev
        for i in range(n - 2, -1, -1):
            prevPotionDone -= skill[i + 1] * mana[j - 1]
            prev = max(prevPotionDone, prev - skill[i] * mana[j])
        prev += sumSkill * mana[j]
    return prev
```

```python
skill = [1,5,2,4]
mana = [5,1,4,2]
minTime(skill, mana)
```

```python
skill = [1,1,1]
mana = [1,1,1]
minTime(skill, mana)
```

```python
skill = [1,2,3,4]
mana = [1,2]
minTime(skill, mana)
```
