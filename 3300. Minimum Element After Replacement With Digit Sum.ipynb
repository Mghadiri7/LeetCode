## First solution

```python
def minElement(nums):
    min_sum = float('inf')
    for n in nums:
        _sum = 0
        while n > 0:
            _sum += (n % 10)
            n //= 10
        if _sum < min_sum:
            min_sum = _sum
    return min_sum
```

```python
nums = [10,12,13,14]
minElement(nums)
```

## Second Solution

```python
def minElement(nums):
    return min([sum([int(s) for s in str(n)]) for n in nums])
```

```python
nums = [999,19,199]
minElement(nums)
```
