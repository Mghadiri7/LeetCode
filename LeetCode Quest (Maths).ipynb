# Arithmetic & Basic Reasoning

### Q1. Can Make Arithmetic Progression From Sequence

```python
def canMakeArithmeticProgression(arr):
    arr.sort()
    n = len(arr) - 1
    diff = arr[0] - arr[1]
    for i in range(1, n):
        if arr[i]-arr[i+1] != diff:
            return False
    return True
```

```python
arr = [3,5,1]
canMakeArithmeticProgression(arr)
```

### Q2. Find the Pivot Integer

```python
def pivotInteger(n):
    _sum = n * (n+1) // 2
    for i in range(n):
        left = i * (i+1) // 2
        right = _sum - left - i - 1
        if left == right:
            return i+1
    else:
        return -1
```

```python
n = 8
pivotInteger(n)
```

### Q3. Palindrome Number

```python
def isPalindrome(x):
        if x<0:
            return False
        res = 0
        temp = x
        while temp>0:
            res = res*10 + temp%10
            temp//=10
        return res==x
```

```python
x = 1221
isPalindrome(x)
```

# Divisibility & Modular Arithmetic

### Q1. Ugly Number

```python
def isUgly(n):
    if n<1:
        return False
    while n%2==0:
        n//=2
    while n%3==0:
        n//=3
    while n%5==0:
        n//=5
    return n==1
```

```python
n = 14
isUgly(n)
```

### Q2. Smallest Integer Divisible by K

```python
def smallestRepunitDivByK(k):
    if k%10 not in {1,3,7,9}:
        return -1

    _mod, mod_set = 0, set()
    for i in range(1, k+1):
        _mod = (10 * _mod + 1) % k
        if _mod==0:
            return i
        if _mod in mod_set:
            return -1
        mod_set.add(_mod)
    return -1
```

```python
k = 3
smallestRepunitDivByK(k)
```

### Q3. Self Dividing Numbers

```python
def selfDividingNumbers(left, right):
    right+=1
    ans = []
    for i in range(left, right):
        num = i
        while i!=0:
            if i%10==0 or num % (i%10)!=0:
                break
            i //= 10
        if i==0:
            ans.append(num)  
    return ans
```

```python
left = 1
right = 22
selfDividingNumbers(left, right)
```

# Combinatorics & Permutations

### Q1. Permutation Sequence

```python
from math import factorial

def getPermutation(n, k):
    k -= 1
    lst = [1,2,3,4,5,6,7,8,9]
    ans = ""
    f = factorial(n-1)
    for i in range(n):
        num = k//f
        ans += str(lst[num])
        del lst[num]
        k %= f
        if i<n-1:
            f //= (n-i-1)
    return ans
```

```python
n = 4
k = 9
getPermutation(n, k)
```

### Q2. Pascal's Triangle

```python
def generate(numRows):
    ans = [[1]]
    for i in range(1, numRows):
        lst = [1]
        for j in range(i-1):
            lst.append(ans[i-1][j] + ans[i-1][j+1])
        lst.append(1)
        ans.append(lst)
    return ans
```

```python
numRows = 5
generate(numRows)
```

### Q3. Number of Ways to Rearrange Sticks With K Sticks Visible

```python
def rearrangeSticks(n, k):
    mod = 10**9+7
    lst = [[0]*(k+1) for i in range(n+1)]
    lst[0][0] = 1
    for i in range(1, n+1):
        for j in range(1, k+1):
            lst[i][j] = (lst[i-1][j-1] + lst[i-1][j] * (i-1)) % mod
    return lst[n][k]
```

```python
n = 20
k = 11
rearrangeSticks(n, k)
```

# Prime numbers, factors & number theory

### Q1. Count Primes

```python
def countPrimes(n):
    seen = [0] * n
    ans = 0
    for i in range(2, n):
        if seen[i]==0:
            ans+=1
            for j in range(i*2, n, i):
                seen[j] = 1
    return ans
```

```python
n = 10
countPrimes(n)
```

### Q2. Factorial Trailing Zeroes

```python
def trailingZeroes(n):
    x = 5
    ans = 0
    while x<=n:
        ans += n//x
        x *= 5
    return ans
```

```python
n = 26
trailingZeroes(n)
```

### Q3. Closest Prime Numbers in Range

```python
def closestPrimes(left, right):
    if right<=2:
        return [-1,-1]

    right += 1
    seen = [1] * (right)
    seen[0] = seen[1] = 0
    for i in range(2, right):
        if seen[i]==1:
            for j in range(i*2, right, i):
                seen[j] = 0

    prev = -1
    _min = 1000001
    ans = [-1, -1]

    for i in range(left, right):
        if seen[i] == 1:
            if prev != -1:
                dist = i - prev
                if dist < _min:
                    _min = dist
                    ans = [prev, i]
            prev = i
    return ans
```

```python
left = 10
right = 19
closestPrimes(left, right)
```

# Geometry

### Q1. Projection Area of 3D Shapes

```python
def projectionArea(grid):
    n = len(grid)
    area = 0
    for row in grid:
        area += max(row)
        for num in row:
            if num>0:
                area += 1
    for col in range(n):
        column = [row[col] for row in grid]
        area += max(column)
    return area
```

```python
grid = [[1,2],[3,4]]
projectionArea(grid)
```

### Q2. Largest Triangle Area

```python
def largestTriangleArea(points):
    length = len(points)
    max_area = 0
    for i in range(length-2):
        for j in range(i+1, length-1):
            for k in range(j+1, length):
                x1, y1 = points[i]
                x2, y2 = points[j]
                x3, y3 = points[k]
                area = abs(0.5 * (x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)))
                max_area = max(max_area, area)
    return max_area
```

```python
points = [[0,0],[0,1],[1,0],[0,2],[2,0]]
largestTriangleArea(points)
```

### Q3. Max Points on a Line

```python
from collections import defaultdict

def maxPoints(points):
    n = len(points)
    if n<3:
        return n

    def find_slope(p1, p2):
        x1, y1 = p1
        x2, y2 = p2
        delta_x = x1-x2
        if delta_x==0:
            return float('inf')
        return (y1-y2) / delta_x
        
    ans = 1
    for i, p1 in enumerate(points):
        slopes = defaultdict(int)
        for p2 in points[i+1:]:
            slope = find_slope(p1, p2)
            slopes[slope]+=1
        if slopes:
            ans = max(max(slopes.values()), ans)

    return ans+1
```

```python
points = [[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]
maxPoints(points)
```

### Q4. Rectangle Area

```python
def computeArea(ax1, ay1, ax2, ay2, bx1, by1, bx2, by2):
    area = (ax2-ax1) * (ay2-ay1) + (bx2-bx1) * (by2-by1)
    if bx1>=ax2 or ax1>=bx2 or by1>=ay2 or ay1>=by2:
        return area
    overlap = min(bx2-ax1, ax2-bx1, ax2-ax1, bx2-bx1) * min(by2-ay1, ay2-by1, ay2-ay1, by2-by1)
    return area-overlap
```

```python
# ax1 = -3, ay1 = 0, ax2 = 3, ay2 = 4, bx1 = 0, by1 = -1, bx2 = 9, by2 = 2
computeArea(-3, 0, 3, 4, 0, -1, 9, 2)
```

# Basic Bit Operations

### Q1. Number of 1 Bits

```python
def hammingWeight(n):
    ans = 0
    while n>0:
        if n%2==1:
            ans += 1
        n //= 2
    return ans
```

```python
n = 11
hammingWeight(n)
```

### Q2. Single Number

```python
def singleNumber(nums):
    res = 0
    for num in nums:
        res ^= num
    return res
```

```python
nums = [4,1,2,1,2]
singleNumber(nums)
```

### Q3. Sum of Two Integers

```python
def getSum(a, b):
    MASK = 0xFFFFFFFF 
    MAX_INT = 0x7FFFFFFF 

    while b != 0:
        carry = (a & b) << 1
        a = a ^ b
        b = carry
        a = a & MASK
        b = b & MASK

    if a > MAX_INT:
        return ~(a ^ MASK)
    else:
        return a
```

```python
a = 2
b = 3
getSum(a, b)
```

# Bitmasking for Sets/states

### Q1. Gray Code

```python
def grayCode(n):
    total_number = 1<<n
    result = []
    for i in range(total_number):
        shifted_i = i>>1
        gray_val = i ^ shifted_i
        result.append(gray_val)
    return result
```

```python
n = 2
grayCode(n)
```

### Q2. Number of Digit One

```python
def countDigitOne(n):
    i = 1
    ans = 0
    while i<=n:
        divider = i * 10
        full_block = (n // divider) * i
        reminder = n % divider
        partial_block = min(max(reminder-i+1, 0), i)
        ans += full_block + partial_block
        i *= 10
    return ans
```

```python
n = 13
countDigitOne(n)
```

### Q3. Single Number II

```python
def singleNumber(nums):
    ones = 0
    twos = 0
    for num in nums:
        ones ^= (num & ~twos)
        twos ^= (num & ~ones)
    return ones
```

```python
nums = [0,1,0,1,0,1,99]
singleNumber(nums)
```
