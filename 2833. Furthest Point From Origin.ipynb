## First Solution

```python
from collections import Counter

def furthestDistanceFromOrigin(moves):
    count = Counter(moves)
    return count['_'] + abs(count['L'] - count['R'])
```

```python
moves = "L_RL__R"
furthestDistanceFromOrigin(moves)
```

```python
moves = "_______"
furthestDistanceFromOrigin(moves)
```

## Second Solution

```python
def furthestDistanceFromOrigin(moves):
    counter = 0
    ans = 0
    for m in moves:
        if m == 'L':
            counter += 1
        elif m == 'R':
            counter -= 1
        else:
            ans += 1
    return ans + abs(counter)
```

```python
moves = "L_RL__R"
furthestDistanceFromOrigin(moves)
```

```python
moves = "_______"
furthestDistanceFromOrigin(moves)
```
