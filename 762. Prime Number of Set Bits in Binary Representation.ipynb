```python
def countPrimeSetBits(left, right):
    count = 0
    for i in range(left, right+1):
        if isPrime(bin(i).count('1')):
            count += 1
    return count


def isPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
left = 10
right = 15
countPrimeSetBits(left, right)
```
