```python
def kLengthApart(nums, k):
    n = len(nums)
    k = k+1
    j = -k
    for i in range(n):
        if nums[i]==1:
            if i-j<k:
                return False
            j = i
    return True
```

```python
nums = [1,0,0,0,1,0,0,1]
k = 2
kLengthApart(nums, k)
```
