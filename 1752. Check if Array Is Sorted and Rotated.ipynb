## First Solution

```python
def check(nums):
    n = len(nums) - 1
    flag = True
    smallest = nums[0]
    for i in range(n):
        if nums[i] <= nums[i+1] and flag == True:
            continue
        elif flag == True and nums[i+1] <= smallest:
            flag = False
            continue
        if nums[i] <= nums[i+1] and nums[i+1] <= smallest:
            continue
        else:
            return False
    return True
```

```python
nums = [3,4,5,1,2]
check(nums)
```

```python
nums = [2,1,3,4]
check(nums)
```

```python
nums = [8,5,4,5,1,4,5,2,2]
check(nums)
```

```python
nums = [1,2,4,5,3]
check(nums)
```

## Second Solution

```python
def check(nums):
    n = len(nums)
    breaks = 0
    for i in range(n):
        if nums[i] > nums[(i+1)%n]:
            breaks += 1
    return breaks <= 1
```

```python
nums = [8,5,4,5,1,4,5,2,2]
check(nums)
```
