```python
def maxProfit(prices, strategy, k):
    n = len(prices)
    lst = []
    for i in range(n):
        lst.append(prices[i] * strategy[i])
    profit = sum(lst)
    half = k//2
    old = sum(lst[:k])
    new = sum(prices[half:k])
    max_diff = max(0, new-old)

    for i in range(k, n):
        j = i - k
        old += lst[i] - lst[j]
        new = new + prices[i] - prices[j + half]
        max_diff = max(max_diff, new-old)
        
    return profit + max_diff
```

```python
prices = [4,2,8]
strategy = [-1,0,1]
k = 2
maxProfit(prices, strategy, k)
```
