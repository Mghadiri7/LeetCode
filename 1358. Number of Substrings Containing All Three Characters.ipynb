```python
def numberOfSubstrings(s):
    last_seen = {'a':-1, 'b':-1, 'c':-1}
    ans = 0
    for i, char in enumerate(s):
        last_seen[char] = i
        min_index = min(last_seen['a'], last_seen['b'], last_seen['c'])

        if min_index != -1:
            ans += (min_index + 1)
            
    return ans
```

```python
s = "abcabc"
numberOfSubstrings(s)
```
